```python
import pandas as pd

whitesox_pitching = pd.read_csv("whitesox_value_pitching_clean.csv")
whitesox_fielding = pd.read_csv("whitesox_standard_fielding_clean.csv")
whitesox_batting = pd.read_csv("whitesox_standard_batting_clean.csv")
cubs_pitching = pd.read_csv("cubs_standard_pitching_clean.csv")
cubs_fielding = pd.read_csv("cubs_standard_fielding_clean.csv")
cubs_batting = pd.read_csv("cubs_standard_batting_clean.csv")

whitesox_pitching.info(), whitesox_fielding.info(), whitesox_batting.info(), cubs_pitching.info(), cubs_fielding.info(), cubs_batting.info()
```

```python
whitesox_batting_avg = whitesox_batting[['OPS', 'rOBA']].mean()
cubs_batting_avg = cubs_batting[['OPS', 'rOBA']].mean()

whitesox_fielding_avg = whitesox_fielding[['Rdrs']].mean()
cubs_fielding_avg = cubs_fielding[['Rdrs']].mean()

whitesox_pitching_avg = whitesox_pitching[['RA9', 'WAR']].mean()
cubs_pitching_avg = cubs_pitching[['ERA', 'WHIP']].mean()

comparison_df = pd.DataFrame({
    'White Sox': [
        whitesox_batting_avg['OPS'], whitesox_batting_avg['rOBA'],
        whitesox_fielding_avg['Rdrs'],
        whitesox_pitching_avg['RA9'], whitesox_pitching_avg['WAR']
    ],
    'Cubs': [
        cubs_batting_avg['OPS'], cubs_batting_avg['rOBA'],
        cubs_fielding_avg['Rdrs'],
        cubs_pitching_avg['ERA'], cubs_pitching_avg['WHIP']
    ]
}, index=['OPS (Batting)', 'rOBA (Batting)', 'Rdrs (Fielding)', 'RA9/ERA (Pitching)', 'WAR/WHIP (Pitching)'])


comparison_df
```

```python
import numpy as np

n_simulations = 10000

white_sox_score = (
    comparison_df.loc['OPS (Batting)', 'White Sox'] +
    comparison_df.loc['rOBA (Batting)', 'White Sox'] +
    comparison_df.loc['Rdrs (Fielding)', 'White Sox'] +
    comparison_df.loc['WAR/WHIP (Pitching)', 'White Sox'] +
    comparison_df.loc['RA9/ERA (Pitching)', 'White Sox']
)

cubs_score = (
    comparison_df.loc['OPS (Batting)', 'Cubs'] +
    comparison_df.loc['rOBA (Batting)', 'Cubs'] +
    comparison_df.loc['Rdrs (Fielding)', 'Cubs'] +
    comparison_df.loc['WAR/WHIP (Pitching)', 'Cubs'] +
    comparison_df.loc['RA9/ERA (Pitching)', 'Cubs']
)

white_sox_results = white_sox_score + np.random.normal(0, 1, n_simulations)
cubs_results = cubs_score + np.random.normal(0, 1, n_simulations)

white_sox_wins = np.sum(white_sox_results > cubs_results)
cubs_wins = np.sum(cubs_results > white_sox_results)
draws = n_simulations - white_sox_wins - cubs_wins

white_sox_win_pct = white_sox_wins / n_simulations
cubs_win_pct = cubs_wins / n_simulations
draw_pct = draws / n_simulations

white_sox_win_pct, cubs_win_pct, draw_pct
```
